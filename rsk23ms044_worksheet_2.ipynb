```python
#Question 1
maxiter = 100
eps = 1.0E-5
import numpy as np
import matplotlib.pyplot as plt
def bisection(f,Left,Right,eps,maxiter):
    if abs(f(Left)) < eps:
        return L,0
    if abs(f(Right)) < eps:
        return R,0
    if f(Left)*f(Right)>0:
        print("Bracketing interval is improper.")
        return None,0
        
    for i in range(maxiter):
        mid=0.5*(Left+Right)
        if abs(f(mid))<eps:
            return mid, i+1
        if f(Left)*f(mid)<0:
            Right = mid
        else:
            Left = mid
    return mid, maxiter

```

```python
def secant(f, x0, x1, eps, maxiter):
    for i in range(maxiter):
        f0 = f(x0)
        f1 = f(x1)
        
        if abs(f1) < eps:
            return x1, i + 1
        
        if abs(f1 - f0) < 1e-12:
            return x1, i + 1
        
        x_new = x1 - f1 * (x1 - x0) / (f1 - f0)
        x0, x1 = x1, x_new
    
    return x1, maxiter
```

```python
def newton_raphson(f, df, x0, eps, maxiter):
    for i in range(maxiter):
        fx = f(x0)
        
        if abs(fx) < eps:
            return x0, i + 1
        
        dfx = df(x0)
        if abs(dfx) < 1e-12:
            print("Error: Derivative too small")
            return x0, i + 1
        
        x0 = x0 - fx / dfx
    
    return x0, maxiter
```

```python
# Define function for Question 1: tan(x) - 3x = 0
def f1(x):
    return np.tan(x) - 3 * x

def df1(x):
    return 1 / (np.cos(x)**2) - 3
```

```python
# Plot the function
x = np.linspace(0.0, 1.57, 1000)
plt.figure(figsize=(8, 5))
plt.plot(x, f1(x), 'b-', linewidth=2)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel("x")
plt.ylabel("tan(x) - 3x")
plt.title("Question 1: 3x = tan(x)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Solve using all three methods
print("Question 1: Solve 3x = tan(x)")
print("=" * 50)

root_bis, iter_bis = bisection(f1, 1.2, 1.4, eps, maxiter)
print(f"Bisection: Root = {root_bis:.8f}, Iterations = {iter_bis}")

root_sec, iter_sec = secant(f1, 1.2, 1.25, eps, maxiter)
print(f"Secant: Root = {root_sec:.8f}, Iterations = {iter_sec}")

root_nr, iter_nr = newton_raphson(f1, df1, 1.2, eps, maxiter)
print(f"Newton-Raphson: Root = {root_nr:.8f}, Iterations = {iter_nr}")
```

```python
#Question 2
def f2(x):
    """V(x) - E = x^4 + (2x/3)^3 - x^2 + 0.125"""
    return x**4 + ((2 * x / 3)**3) - x**2 + 0.125

# Plot the potential
x = np.linspace(-1.25, 1.25, 1000)
plt.figure(figsize=(8, 5))
plt.plot(x, f2(x), 'g-', linewidth=2)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel("x")
plt.ylabel("V(x) - E")
plt.title("Question 2: Double-Well Potential")
plt.grid(True, alpha=0.3)
plt.show()

print("\nQuestion 2: Turning points in double-well potential")
print("=" * 50)
```

```python
# (a) Bisection method
print("(a) Bisection Method:")

print("Turning point 1:")
tp1_bis, iter1_bis = bisection(f2, -1.25, -0.5, eps, maxiter)
if tp1_bis is not None:
    print(f"  Root = {tp1_bis:.8f}, Iterations = {iter1_bis}")

print("Turning point 2:")
tp2_bis, iter2_bis = bisection(f2, -0.5, 0, eps, maxiter)
if tp2_bis is not None:
    print(f"  Root = {tp2_bis:.8f}, Iterations = {iter2_bis}")

print("Turning point 3:")
tp3_bis, iter3_bis = bisection(f2, 0.25, 0.5, eps, maxiter)
if tp3_bis is not None:
    print(f"  Root = {tp3_bis:.8f}, Iterations = {iter3_bis}")

print("Turning point 4:")
tp4_bis, iter4_bis = bisection(f2, 0.5, 1, eps, maxiter)
if tp4_bis is not None:
    print(f"  Root = {tp4_bis:.8f}, Iterations = {iter4_bis}")
```

```python
# (b) Secant method
print("\n(b) Secant Method:")

print("Turning point 1:")
tp1_sec, iter1_sec = secant(f2, -1.25, -1, eps, maxiter)
if tp1_sec is not None:
    print(f"  Root = {tp1_sec:.8f}, Iterations = {iter1_sec}")

print("Turning point 2:")
tp2_sec, iter2_sec = secant(f2, -0.5, 0, eps, maxiter)
if tp2_sec is not None:
    print(f"  Root = {tp2_sec:.8f}, Iterations = {iter2_sec}")

print("Turning point 3:")
tp3_sec, iter3_sec = secant(f2, 0.25, 0.5, eps, maxiter)
if tp3_sec is not None:
    print(f"  Root = {tp3_sec:.8f}, Iterations = {iter3_sec}")

print("Turning point 4:")
tp4_sec, iter4_sec = secant(f2, 0.7, 1, eps, maxiter)
if tp4_sec is not None:
    print(f"  Root = {tp4_sec:.8f}, Iterations = {iter4_sec}")

print("\n" + "=" * 50)
```
