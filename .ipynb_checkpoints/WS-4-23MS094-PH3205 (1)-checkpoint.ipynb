```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Helper functions for numerical methods

def euler_method(f, x0, t_array):
    """Euler method for dx/dt = f(x, t)"""
    x_array = np.zeros(len(t_array))
    x_array[0] = x0
    for i in range(len(t_array) - 1):
        h = t_array[i+1] - t_array[i]
        x_array[i+1] = x_array[i] + h * f(x_array[i], t_array[i])
    return x_array

def midpoint_method(f, x0, t_array):
    """Midpoint method for dx/dt = f(x, t)"""
    x_array = np.zeros(len(t_array))
    x_array[0] = x0
    for i in range(len(t_array) - 1):
        h = t_array[i+1] - t_array[i]
        t_mid = t_array[i] + h/2
        x_mid = x_array[i] + (h/2) * f(x_array[i], t_array[i])
        x_array[i+1] = x_array[i] + h * f(x_mid, t_mid)
    return x_array

def euler_method_coupled(f, state0, t_array):
    """Euler method for coupled ODEs"""
    n_points = len(t_array)
    n_vars = len(state0)
    state_array = np.zeros((n_points, n_vars))
    state_array[0] = state0
    for i in range(n_points - 1):
        h = t_array[i+1] - t_array[i]
        state_array[i+1] = state_array[i] + h * f(state_array[i], t_array[i])
    return state_array

def midpoint_method_coupled(f, state0, t_array):
    """Midpoint method for coupled ODEs"""
    n_points = len(t_array)
    n_vars = len(state0)
    state_array = np.zeros((n_points, n_vars))
    state_array[0] = state0
    for i in range(n_points - 1):
        h = t_array[i+1] - t_array[i]
        t_mid = t_array[i] + h/2
        state_mid = state_array[i] + (h/2) * f(state_array[i], t_array[i])
        state_array[i+1] = state_array[i] + h * f(state_mid, t_mid)
    return state_array
```

```python
# Problem 1: dx/dt = -xt

def problem1_derivative(x, t):
    return -x * t

def exact_solution_problem1(t):
    return np.exp(-t**2 / 2)
```

```python
# Question 1(a): Solve with h = 0.01 from t = 0 to t = 15

h1a = 0.01
tini = 0.0
tfinal = 15.0
x0_p1 = 1.0

t_array_1a = np.arange(tini, tfinal + h1a, h1a)
x_euler_1a = euler_method(problem1_derivative, x0_p1, t_array_1a)
x_midpoint_1a = midpoint_method(problem1_derivative, x0_p1, t_array_1a)
x_exact_1a = exact_solution_problem1(t_array_1a)

plt.figure(figsize=(10, 6))
plt.plot(t_array_1a, x_exact_1a, 'k-', linewidth=2, label='Exact Solution')
plt.plot(t_array_1a, x_euler_1a, 'r--', linewidth=1.5, label='Euler Method')
plt.plot(t_array_1a, x_midpoint_1a, 'b:', linewidth=1.5, label='Midpoint Method')
plt.xlabel('t', fontsize=12)
plt.ylabel('x(t)', fontsize=12)
plt.title('Problem 1(a): Solutions for dx/dt = -xt (h = 0.01)', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Question 1(b): Error analysis for varying h

n_values = np.arange(-4, -1.8, 0.2)
h_values = 10.0 ** n_values

errors_euler_p1 = []
errors_midpoint_p1 = []

for h in h_values:
    t_array = np.arange(tini, tfinal + h, h)
    
    x_euler = euler_method(problem1_derivative, x0_p1, t_array)
    x_exact = exact_solution_problem1(t_array[-1])
    error_euler = abs(x_euler[-1] - x_exact)
    errors_euler_p1.append(error_euler)
    
    x_midpoint = midpoint_method(problem1_derivative, x0_p1, t_array)
    error_midpoint = abs(x_midpoint[-1] - x_exact)
    errors_midpoint_p1.append(error_midpoint)
```

```python
# Question 1(c): Plot and fit log10(h) vs log10(error)

log_h = np.log10(h_values)
log_errors_euler_p1 = np.log10(errors_euler_p1)
log_errors_midpoint_p1 = np.log10(errors_midpoint_p1)

def linear_fit(x, slope, intercept):
    return slope * x + intercept

params_euler_p1, _ = curve_fit(linear_fit, log_h, log_errors_euler_p1)
slope_euler_p1, intercept_euler_p1 = params_euler_p1

params_midpoint_p1, _ = curve_fit(linear_fit, log_h, log_errors_midpoint_p1)
slope_midpoint_p1, intercept_midpoint_p1 = params_midpoint_p1

plt.figure(figsize=(10, 6))
plt.plot(log_h, log_errors_euler_p1, 'ro', markersize=8, label='Euler Method')
plt.plot(log_h, linear_fit(log_h, *params_euler_p1), 'r--', 
         linewidth=2, label=f'Euler Fit: slope = {slope_euler_p1:.3f}')
plt.plot(log_h, log_errors_midpoint_p1, 'bs', markersize=8, label='Midpoint Method')
plt.plot(log_h, linear_fit(log_h, *params_midpoint_p1), 'b--', 
         linewidth=2, label=f'Midpoint Fit: slope = {slope_midpoint_p1:.3f}')
plt.xlabel('log₁₀(h)', fontsize=12)
plt.ylabel('log₁₀(error)', fontsize=12)
plt.title('Problem 1(c): Error Scaling Analysis', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Euler method: error ∝ h^{slope_euler_p1:.3f}")
print(f"Midpoint method: error ∝ h^{slope_midpoint_p1:.3f}")
```

```python
# Problem 2: Coupled ODEs - Simple Harmonic Motion
# dy/dt = p, dp/dt = -4π²y

def problem2_derivatives(state, t):
    y, p = state
    dydt = p
    dpdt = -4 * np.pi**2 * y
    return np.array([dydt, dpdt])

def exact_solution_problem2(t, y0=1.0, p0=0.0):
    omega = 2 * np.pi
    y = y0 * np.cos(omega * t) + (p0 / omega) * np.sin(omega * t)
    p = -y0 * omega * np.sin(omega * t) + p0 * np.cos(omega * t)
    return y, p
```

```python
# Question 2(a): Solve with h = 0.01 from t = 0 to t = 15

h2a = 0.01
state0_p2 = np.array([1.0, 0.0])

t_array_2a = np.arange(tini, tfinal + h2a, h2a)
state_euler_2a = euler_method_coupled(problem2_derivatives, state0_p2, t_array_2a)
state_midpoint_2a = midpoint_method_coupled(problem2_derivatives, state0_p2, t_array_2a)
y_exact_2a, p_exact_2a = exact_solution_problem2(t_array_2a)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t_array_2a, y_exact_2a, 'k-', linewidth=2, label='Exact Solution')
plt.plot(t_array_2a, state_euler_2a[:, 0], 'r--', linewidth=1.5, label='Euler Method')
plt.plot(t_array_2a, state_midpoint_2a[:, 0], 'b:', linewidth=1.5, label='Midpoint Method')
plt.xlabel('t', fontsize=12)
plt.ylabel('y(t)', fontsize=12)
plt.title('Position y(t)', fontsize=13)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(t_array_2a, p_exact_2a, 'k-', linewidth=2, label='Exact Solution')
plt.plot(t_array_2a, state_euler_2a[:, 1], 'r--', linewidth=1.5, label='Euler Method')
plt.plot(t_array_2a, state_midpoint_2a[:, 1], 'b:', linewidth=1.5, label='Midpoint Method')
plt.xlabel('t', fontsize=12)
plt.ylabel('p(t)', fontsize=12)
plt.title('Momentum p(t)', fontsize=13)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

plt.suptitle('Problem 2(a): Simple Harmonic Motion (h = 0.01)', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Phase space plot

plt.figure(figsize=(10, 6))
plt.plot(y_exact_2a, p_exact_2a, 'k-', linewidth=2, label='Exact Solution')
plt.plot(state_euler_2a[:, 0], state_euler_2a[:, 1], 'r--', linewidth=1.5, label='Euler Method')
plt.plot(state_midpoint_2a[:, 0], state_midpoint_2a[:, 1], 'b:', linewidth=1.5, label='Midpoint Method')
plt.xlabel('y (position)', fontsize=12)
plt.ylabel('p (momentum)', fontsize=12)
plt.title('Problem 2(a): Phase Space Trajectory', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Question 2(b): Error analysis for varying h

errors_euler_p2 = []
errors_midpoint_p2 = []

for h in h_values:
    t_array = np.arange(tini, tfinal + h, h)
    
    state_euler = euler_method_coupled(problem2_derivatives, state0_p2, t_array)
    y_exact_final, _ = exact_solution_problem2(t_array[-1])
    error_euler = abs(state_euler[-1, 0] - y_exact_final)
    errors_euler_p2.append(error_euler)
    
    state_midpoint = midpoint_method_coupled(problem2_derivatives, state0_p2, t_array)
    error_midpoint = abs(state_midpoint[-1, 0] - y_exact_final)
    errors_midpoint_p2.append(error_midpoint)
```

```python
# Question 2(c): Plot and fit log10(h) vs log10(error)

log_errors_euler_p2 = np.log10(errors_euler_p2)
log_errors_midpoint_p2 = np.log10(errors_midpoint_p2)

params_euler_p2, _ = curve_fit(linear_fit, log_h, log_errors_euler_p2)
slope_euler_p2, intercept_euler_p2 = params_euler_p2

params_midpoint_p2, _ = curve_fit(linear_fit, log_h, log_errors_midpoint_p2)
slope_midpoint_p2, intercept_midpoint_p2 = params_midpoint_p2

plt.figure(figsize=(10, 6))
plt.plot(log_h, log_errors_euler_p2, 'ro', markersize=8, label='Euler Method')
plt.plot(log_h, linear_fit(log_h, *params_euler_p2), 'r--', 
         linewidth=2, label=f'Euler Fit: slope = {slope_euler_p2:.3f}')
plt.plot(log_h, log_errors_midpoint_p2, 'bs', markersize=8, label='Midpoint Method')
plt.plot(log_h, linear_fit(log_h, *params_midpoint_p2), 'b--', 
         linewidth=2, label=f'Midpoint Fit: slope = {slope_midpoint_p2:.3f}')
plt.xlabel('log₁₀(h)', fontsize=12)
plt.ylabel('log₁₀(error)', fontsize=12)
plt.title('Problem 2(c): Error Scaling Analysis', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Euler method: error ∝ h^{slope_euler_p2:.3f}")
print(f"Midpoint method: error ∝ h^{slope_midpoint_p2:.3f}")
```
