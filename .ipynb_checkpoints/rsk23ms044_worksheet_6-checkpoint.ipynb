```python
import numpy as np
import matplotlib.pyplot as plt

# Dormand-Prince coefficients
a2 = 1/5;  b21 = 1/5
a3 = 3/10; b31 = 3/40;       b32 = 9/40
a4 = 4/5;  b41 = 44/45;      b42 = -56/15;      b43 = 32/9
a5 = 8/9;  b51 = 19372/6561; b52 = -25360/2187; b53 = 64448/6561; b54 = -212/729
a6 = 1;    b61 = 9017/3168;  b62 = -355/33;     b63 = 46732/5247; b64 = 49/176;   b65 = -5103/18656
a7 = 1

c1 = 35/384;     c2 = 0; c3 = 500/1113;   c4 = 125/192;  c5 = -2187/6784;    c6 = 11/84;    c7 = 0
c1s = 5179/57600; c2s = 0; c3s = 7571/16695; c4s = 393/640; c5s = -92097/339200; c6s = 187/2100; c7s = 1/40

def rk4(fn, x, y, h):
    k1 = h*fn(x, y)
    k2 = h*fn(x + h/2, y + k1/2)
    k3 = h*fn(x + h/2, y + k2/2)
    k4 = h*fn(x + h, y + k3)
    return y + (k1 + 2*k2 + 2*k3 + k4)/6

def caller(my_method, fn, y_ini, x0, xT, h):
    xs = np.arange(x0, xT, h)
    N = len(xs)
    y = np.asarray(y_ini)
    ys = np.zeros((N, len(y_ini)))
    for i in range(N):
        ys[i, :] = y
        y = my_method(fn, xs[i], y, h)
    return xs, ys

def erk54h(f, x, y, h, k7bh):
    if sum(k7bh) == 0:
        k1 = h*f(x, y)
    else:
        k1 = k7bh*h
    k2 = h*f(x + a2*h, y + b21*k1)
    k3 = h*f(x + a3*h, y + b31*k1 + b32*k2)
    k4 = h*f(x + a4*h, y + b41*k1 + b42*k2 + b43*k3)
    k5 = h*f(x + a5*h, y + b51*k1 + b52*k2 + b53*k3 + b54*k4)
    k6 = h*f(x + a6*h, y + b61*k1 + b62*k2 + b63*k3 + b64*k4 + b65*k5)
    y5 = y + c1*k1 + c2*k2 + c3*k3 + c4*k4 + c5*k5 + c6*k6
    k7bh = f(x + a7*h, y5)
    y4 = y + c1s*k1 + c2s*k2 + c3s*k3 + c4s*k4 + c5s*k5 + c6s*k6 + c7s*k7bh*h
    err = abs(y5 - y4)
    return y5, err, k7bh

def caller54(fn, y_ini, x0, xT, h0, max_iter, abstol, reltol):
    y = np.asarray(y_ini); h = h0
    ys = np.zeros((max_iter, len(y_ini)))
    xs = np.zeros((max_iter, 1))
    xs[0] = x0; ys[0, :] = y
    x = x0; i = 0
    k7bh = np.zeros(len(y_ini))
    while x <= xT and i < max_iter:
        tol = abstol + reltol*np.linalg.norm(ys[i, :])
        y, err, k7bh = erk54h(fn, xs[i], ys[i, :], h, k7bh)
        merr = np.max(err)
        if merr == 0.0:
            merr = tol/100
        hnew = 0.9*h*(tol/merr)**0.2
        if merr < tol and h/5 < hnew < 5*h:
            x += h
            i += 1
            xs[i] = x
            ys[i, :] = y
        h = hnew
    return xs[1:i], ys[1:i, :]
```

```python
# Question 1 setup
mu = 5.0
lam = 40.0

def vdp(x, state):
    y, p = state
    return np.array([p, mu*(1 - y**2)*p - lam*y])

y_ini = np.array([0.5, 0.0])
x0 = 0.0
xT = 20.0
```

```python
# (a) RK4 with h = 1e-4
xs_rk4, ys_rk4 = caller(rk4, vdp, y_ini, x0, xT, 1e-4)

plt.figure(figsize=(10, 5))
plt.plot(xs_rk4, ys_rk4[:, 0], 'b-', linewidth=0.8, label="y(x)")
plt.plot(xs_rk4, ys_rk4[:, 1], 'r-', linewidth=0.8, label="y'(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Van der Pol: RK4, h = 1e-4')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# (b) Dormand-Prince adaptive
xs_dp, ys_dp = caller54(vdp, y_ini, x0, xT, 0.01, 100000, 1e-6, 1e-8)

print(f'Number of adaptive steps: {len(xs_dp)}')

xs_dp_flat = xs_dp[:, 0]

plt.figure(figsize=(10, 5))
plt.plot(xs_dp_flat, ys_dp[:, 0], 'b-', linewidth=0.8, label="y(x)")
plt.plot(xs_dp_flat, ys_dp[:, 1], 'r-', linewidth=0.8, label="y'(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Van der Pol: Dormand-Prince (steps = {len(xs_dp)})')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(np.diff(xs_dp_flat), 'b.', markersize=2)
plt.xlabel('Step number')
plt.ylabel('h')
plt.title('Adaptive step sizes')
plt.grid(alpha=0.3)
plt.show()
```

```python
# (c) Phase space
plt.figure(figsize=(7, 7))
plt.plot(ys_rk4[:, 0], ys_rk4[:, 1], 'b-', linewidth=0.5, label='RK4', alpha=0.7)
plt.plot(ys_dp[:, 0], ys_dp[:, 1], 'r--', linewidth=0.8, label='Dormand-Prince', alpha=0.8)
plt.xlabel('y')
plt.ylabel("y'")
plt.title('Phase Space: Van der Pol Oscillator')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
