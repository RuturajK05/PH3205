```python
maxiter=50 # maximum number of iteration
```

```python
eps=1.0E-5 #epsilon is 10^-5
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#function is 3x=tan(x)
```

```python
def bisection(f,L,R):
    if f(L)==0:
        return L,0
    if f(R)==0:
        return R,0
    if f(L)*f(R)>0:
        return("improper bracketing interval")
    
    for i in range(maxiter):
        M=0.5*(L+R)
        if abs(f(M))<eps:
            return M,i+1
        if f(L)*f(M)<0:
            R=M
        else:
            L=M
    return M,maxiter
```

```python
def secant(f,x1,x2):
    for i in range(maxiter):
        f1,f2=f(x1),f(x2)
        if abs(f1)<eps:
            return x1,i+1
        if abs(f2) <eps:
            return x2,i+1
        if f2-f1==0:
            print("Secant Slope Zero")
            break
        x3=x2-(f2*((x2-x1)/(f2-f1)))
        x2,x1=x3,x2
    return x2,maxiter
```

```python
def NR(f,x1,df): #input the derivative of x as well
    for i in range (maxiter):
        fx=f(x1)
        if abs(fx)<eps:
            return x1,i+1
        dfx=df(x1) 
        if dfx==0:
            print("Newton Raphson doesnt work for derivative 0")
            break
        x1=x1-fx/dfx
    return x1,maxiter
```

```python
def f1(x):
    return 3*x-np.tan(x)
def df1(x):
    return 3-1/np.cos(x)**2
```

```python
x=np.linspace(0.0,1.57,1000)
plt.plot(x,f1(x))
plt.axhline(0)
plt.xlabel("x")
plt.ylabel("3x-tan(x)")
plt.title("3x=tan(x)")
plt.show()
```

```python
bis,iter1=bisection(f1,1.2,1.4)
sec,iter2=secant(f1,1.2,1.25)
nr,iter3=NR(f1,1.2,df1)
```

```python
print("Bisection gives",bis,"in iteration",iter1)
print("Secant gives",sec,"in iteration",iter2)
print("Newton Raphson gives",nr,"in iteration",iter3)
```

```python
def V(x): #Kinetic Energy
    return x**4+((2/3)*x)**3-x**2+0.125
```

```python
x=np.linspace(-1.25,1.25,1000)
plt.plot(x,V(x))
plt.axhline(0)
plt.xlabel("x")
plt.ylabel("V(x)")
plt.title("Potential Well")
plt.show()
```

```python
#well 1
well_bis,iter1=bisection(V,-1.25,-0.5)
well_sec,iter2=secant(V,-1.25,-1)
print("Turning point 1")
print("Bisection gives",well_bis,"in iteration",iter1)
print("Secant gives",well_sec,"in iteration",iter2)
print("Turning point 2")
bis,iter1=bisection(V,-0.5,0)
sec,iter2=secant(V,-0.5,0)
print("Bisection gives",bis,"in iteration",iter1)
print("Secant gives",sec,"in iteration",iter2)
```

```python
#well 2
print("Turning point 3")
well_bis,iter1=bisection(V,0.25,0.5)
well_sec,iter2=secant(V,0.25,0.5)
print("Bisection gives",well_bis,"in iteration",iter1)
print("Secant gives",well_sec,"in iteration",iter2)
print("Turning point 4")
bis,iter1=bisection(V,0.5,1)
sec,iter2=secant(V,0.7,1)
print("Bisection gives",bis,"in iteration",iter1)
print("Secant gives",sec,"in iteration",iter2)
```
