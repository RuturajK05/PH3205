```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#Question 1
```

```python
def forward_diff(f,xn,h):
    return (f(xn+h)-f(xn))/h
def backward_diff(f,xn,h):
    return (f(xn)-f(xn-h))/h
def central_diff(f,xn,h):
    return (f(xn+h)-f(xn-h))/(2*h)
def five_point_approx(f,xn,h):
    return (f(xn-2*h)-8*f(xn-h)+8*f(xn+h)-f(xn+2*h))/(12*h)
```

```python
h_val=[0.5, 0.2, 0.1, 0.05, 0.02, 0.01, 0.005, 0.002, 0.001, 0.0005, 0.0002, 0.0001]
```

```python
def f(x):
    return np.sin(x)
xn=(2/5)*(np.pi)
```

```python
print("H values     Forward          Backward           Central                 Five Pt          Exact")
for h in h_val:
    fd=forward_diff(f,xn,h)
    bd=backward_diff(f,xn,h)
    cd=central_diff(f,xn,h)
    fpa=five_point_approx(f,xn,h)
    exact=np.cos(xn)
    print(h," ",fd," ",bd," ",cd," ",fpa," ",exact,"\n")
```

```python
methods=[forward_diff,backward_diff,central_diff,five_point_approx]
descriptions=["Forward","Backward","Central","Five-point"]
k=0
for j in range(len(methods)):
    ys=[abs(methods[j](f,xn,h)-exact) for h in h_val]
    for i in range(len(ys)):
        print(ys[i],"by",descriptions[j],"method", "with h value",h_val[i])
    plt.plot((h_val),(ys),"o",label=descriptions[k])
    plt.legend()
    k += 1
    
```

```python
def fit_loglog(h,err):
    x=np.log10(h)
    y=np.log10(err)
    p,c=np.polyfit(x,y,1)
    xx=np.linspace(x.min(),x.max(),100)
    yy=p*xx+c
    return p,xx,yy
```

```python
methods=[forward_diff,backward_diff,central_diff,five_point_approx]
labels=["Forward","Backward","Central","Five-point"]
plt.figure()
for k in range(len(methods)):
    err=[abs(methods[k](f,xn,h)-exact) for h in h_val]
    plt.plot(np.log10(h_val),np.log10(err),"p",label=labels[k])
    p,xx,yy=fit_loglog(h_val,err)
    plt.plot(xx,yy,"-",label=labels[k]+" fit (slope="+str(round(p,2))+")")
plt.xlabel("log10(h)")
plt.ylabel("log10(error)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def f(x):
    return np.exp(x)
# exact value will be e-1
exact= np.exp(1)-1
```

```python
def trapezoidal(fx,h):
    return h*(0.5*fx[0]+fx[1:-1].sum()+0.5*fx[-1])
def simpson_1_3(fx,h):
    odd=fx[1:-1:2].sum()
    even=fx[2:-2:2].sum()
    return h/3 * (fx[0]+fx[-1]+4*odd+2*even)

def simpson_3_8(fx,h):
    idx=np.arange(len(fx))
    s3=fx[(idx%3!=0)&(idx!=0)&(idx!=len(fx)-1)].sum()
    s2=fx[(idx% 3 == 0)&(idx != 0)&(idx!=len(fx)-1)].sum()
    return 3*h/8*(fx[0]+fx[-1]+3*s3 + 2*s2)

def boole(fx,h):
    weights=np.zeros_like(fx)
    weights[0]=weights[-1] = 7
    for i in range(1,len(fx)-1):
        r=i%4
        if r==1 or r==3:
            weights[i] = 32
        elif r==2:
            weights[i]=12
        else:
            weights[i]=14
    return 2*h/45*np.sum(weights * fx)
```

```python
M=[4,8,16,32,64]
trap=[]
s13=[]
s38=[]
b=[]
for m in M:
    h=1/(2*m)
    x=np.linspace(0,1,(2*m)+1)
    fx=f(x)
    trap.append(trapezoidal(fx,h))
    s13.append(simpson_1_3(fx,h))
for m in M:
    h=1/(3*m)
    x=np.linspace(0,1,(3*m)+1)
    fx=f(x)
    s38.append(simpson_3_8(fx,h))
for m in M:
    h=1/(4*m)
    x=np.linspace(0,1,(4*m)+1)
    fx=f(x)
    b.append(boole(fx,h))
for i in range(len(M)):
    print("for m value",M[i])
    print(trap[i],"is the answer by trapezoidal method")
    print(s13[i],"is the answer by Simpson's 1/3rd rule")
    print(s38[i],"is the answer by Simpson's 3/8th rule")
    print(b[i],"is the answer by Boole's Rule")
```
