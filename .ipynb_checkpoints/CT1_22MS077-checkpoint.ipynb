```python
import numpy as np
import matplotlib.pyplot as plt
```

Taking $\sigma = \epsilon = 1$,

The expression for force
$$F(r) = -24 \bigg( \frac{1}{r^7} - \frac{2}{r^{13}} \bigg)$$

### Finding root $F(r^*) = 0$

```python
def F(r):
    return -24*(r**(-7) - 2*r**(-13))

# generating radial distances
R = np.linspace(0.8,2,100)

plt.plot(R,F(R),lw=1,c='b')
plt.xlabel('r')
plt.ylabel('F(r)')
```

```python
#=============================
# We first find the interval within which the root lies 

 
#interval size
h = 0.05

x0 = 0.8+h #start of interval
x_e = 2 #end of interval

def tab_method(f,x0,xe,h):
    
    values = np.arange(x0,xe+h,h)
    print('  %s\t    %s\t   %s\t   %s'%('a','b','f(a)','f(b)'))
    for i in range(len(values)-1):
        a= values[i]
        b= values[i+1]
        print('%4.3f\t %4.3f\t %5.3f\t %5.3f'%(a,b,f(a),f(b)))
    
        # whenever sign changes, means the function has crossed X axis
        if f(a)*f(b) <= 0:
            return [round(a,2),round(b,2)]
        
intvl = tab_method(F,x0,x_e,h)
print('The root lies in the interval ',intvl)

```

### Secant Method

```python
def secant_method(f,intvl,err=10**(-5),maxIter=20):
    # err is the tolerance
    #initial guesses
    x0 = intvl[0]
    x1=intvl[1]
    
    eps = 10e-4
    # small threshold alert
    f0 = f(x0)
    f1 = f(x1)
    
    if np.abs(f0) < err:
        return x0
    elif np.abs(f1) < err:
        return x1
    elif np.abs(f0-f1) < eps:
        print('Flat Function. May not converge')
        return 'Invalid Interval'
    else:
        for steps in range(maxIter):
            x2 = x1 - f(x1)*(x1-x0) / (f(x1) - f(x0))
            if (np.abs(f(x2)) < err) or (np.abs(x2-x1) < err):
                # if value is less than tolerance, then x2 is ROOT!
                return x2
            #otherwise
            elif np.abs(f(x0)-f(x1)) < eps:
                print('Flat Function. May not converge')
                break
            else:
                x0 = x1
                x1 = x2
    if ~(np.abs(f(x2)) < err) or ~(np.abs(x2-x1) < err):
        #if both conditions are false at the end
        return 'Did Not Converge!'

rb = secant_method(F,intvl)
print(f'The equilibrium bond distance is computed to be r* = {rb:.10f}')
print(f'The error from analytical result 2^(1/6) is = {np.abs(rb-2**(1/6)):.10f}')
```

```python
def f(x):
    res = (x**3)/(np.exp(x) - 1)
    return res
    
```

```python
## USING BODES/QUARTIC RULE 
def Quartic(func,intvl):
    #
    #interval endpoints
    a = intvl[0]
    b = intvl[1]
    # multiples of 3n+1
    points = [21, 201, 2001, 20001] #N
    area = np.array([0.,0.,0.,0.])
    steps = np.array([0.,0.,0.,0.])
    
    for i in range(4):
        h = (b-a)/(points[i]-1) #step sizes
        steps[i] += h
#         print(h)
        xs = np.arange(a,b+h,h)
        
        area[i] += (2*h/45)*(7*np.sum(func(xs)[::4]) + 32*np.sum(func(xs)[1:-1:4]) + 12*np.sum(func(xs)[2:-1:4]) + 32*np.sum(func(xs)[3:-1:4]) + 7*np.sum(func(xs)[4:-1:4]))
        
  
    
    return area,steps
```

```python
## as 0 is singularity we take a small a value near zero and similarly a very large value like 100 to approximate infinity,as the curve dies down quickly
#xmax = 100
intvl = [0.1,100] #interval
xs = np.linspace(intvl[0], intvl[1],1000)
plt.plot(xs,f(xs),lw=1,c='g')
```

```python
int_Area, stepSizes = Quartic(f,intvl)
int_Area
```

```python
# we find the error from each step size 
absErr = np.abs(int_Area - (np.pi**4)/(15))
```

```python
print('\t Numerical Integration Bodes Method')
print('  %2s\t|      %6s\t \t\t  %10s\t'%('h','Result','Abs Error'))
print('.'*50)
for i in range(len(stepSizes)):
    print('%1.4f\t| %1.15f\t| %1.15f\t|'%(stepSizes[i],int_Area[i],absErr[i]))
```

So, we can conclude a smaller step size, enables the numerical integration to converge nicely

```python
#initial conditions
m = 1 #kg
v0 = 10 #m/s
theta0 = np.deg2rad(60) 
g = 9.8 #m/s^2 acceleration due to gravity
Cd = 1 #kg/s
```

```python
def caller_nonsimplectic(my_method, func, Y_init,N, xs, h):
    y = Y_init
    ys = np.zeros((N, len(Y_init)))
    for i in range(N):
        x = xs[i]
        ys[i, :] = y
        y = my_method(func, x, y, h)
    return ys


def rk4(f1, x, y, h):
    k1 = h * f1(x, y)
    k2 = h * f1(x + h / 2, y + k1 / 2)
    k3 = h * f1(x + h / 2, y + k2 / 2)
    k4 = h * f1(x + h, y + k3)
    return y + (k1 + 2 * k2 + 2 * k3 + k4) / 6
```

The system of ode are

$$ \frac{dx}{dt} = v_x$$
$$ \frac{dy}{dt} = v_y$$
$$ \frac{dv_x}{dt} = -\frac{C_d}{m} v_x \sqrt{v_x^2 + v_y^2}$$
$$ \frac{dv_y}{dt} = -g -\frac{C_d}{m} v_y \sqrt{v_x^2 + v_y^2}$$

Initial value conditions:
$\tan{(\theta_0)} = \frac{v_y(0)}{v_x(0)}$, $v_x(0)^2 + v_y(0)^2 = 10$

$x(0) = y(0) = 0$


Defining a single vector $\vec{A} = [x,v_x,y,v_y]$


To solve $\frac{d \vec{A}}{dt} = \vec{B}(x,v_x,y,v_y)$


We are taking $\Delta t = 0.005$

```python
def B(t,A):
    vector = np.asarray([A[1], -Cd*A[1]*np.sqrt(A[1]**2 + A[3]**2), A[3], -9.8 - Cd*A[3]*np.sqrt(A[1]**2 + A[3]**2)])
    return vector
```

```python
# we first solve for the number of steps required to hit ground, i.e y =0 again

#time step size 
dt = 0.005
#initial value array
A_init = [0.,v0*np.cos(theta0),0.0,v0*np.sin(theta0)] ##[x0,vx0,y0,vy0]
# print(A_init)
t0= 0.0
A_init = rk4(B,t0,A_init,dt)
# print(A_init)
N = 1
while(A_init[2]>0): #i.e till y coordinate is positive
    t0 = t0+dt
    A_init = rk4(B,t0,A_init,dt)
    N+=1

print(N)
```

```python
N = 177+1
ts = np.arange(t0, t0 + N*dt, dt)
#initial value array
A_init = [0.,v0*np.cos(theta0),0.0,v0*np.sin(theta0)] ##[x0,vx0,y0,vy0]
# solving the ODEs
phaseSpace = caller_nonsimplectic(rk4,B,A_init,N,ts,dt)
# this is a array of [x,vx,y,vy] vectors
```

```python
plt.figure(figsize=(10,6))
plt.plot(phaseSpace[:,0], phaseSpace[:,2],lw=2,c='g',label='$C_d = 1$')
plt.title("RK4 Method Trajectory")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.legend()
plt.grid()
plt.plot()
```

# Range

```python
#range (the final x position when y = 0 again).

rangePr = phaseSpace[-1,0]
print(f'The range of the projectile for v0 = {v0}m/s, angle = {np.round(np.rad2deg(theta0),1)} deg, Cd = {Cd} is \n = {rangePr:.4f} m')
```

## Variation with $C_d$

```python
plt.figure(figsize=(12, 18))
plt.title(r" Trajectory variation with $C_d$ (RK4 Method)")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.grid()

cds = np.arange(0,14,2)
# projectile ranges
rangeCds = []

for Cd in cds:
    # getting N
    #time step size 
    dt = 0.005
    #initial value array
    A_init = [0.,v0*np.cos(theta0),0.0,v0*np.sin(theta0)] ##[x0,vx0,y0,vy0]
    t0= 0.0
    A_init = rk4(B,t0,A_init,dt)
    N = 1
    while(A_init[2]>0): #i.e till y coordinate is positive
        t0 = t0+dt
        A_init = rk4(B,t0,A_init,dt)
        N+=1

    N +=1 # just when y value is negative
    ts = np.arange(t0, t0 + N*dt, dt)
    #initial value array
    A_init = [0.,v0*np.cos(theta0),0.0,v0*np.sin(theta0)] ##[x0,vx0,y0,vy0]
    # solving the ODEs
    phaseSpace = caller_nonsimplectic(rk4,B,A_init,N,ts,dt)
    rangeCds.append(phaseSpace[-1,0])
    # this is a array of [x,vx,y,vy] vectors
    plt.plot(phaseSpace[:,0], phaseSpace[:,2], label=r'$C_d =$'+str(Cd),lw=1)
    plt.legend()
    # plt.plot()
```

```python
plt.figure(figsize=(10,7))
plt.plot(cds,rangeCds,lw=1,c='r')
plt.xlabel(r'Drag Coefficient $C_d$')
plt.ylabel('Range (m)')
plt.grid()
plt.title('Variation of Range with drag coefficent')
```

```python
print('\tVariation of Range with drag coefficent')
print()
print('  %3s\t|   %6s\t'%('C_d','Range'))
print('.'*25)
for i in range(len(cds)):
    print('%1.1f\t| %1.4f\t|'%(cds[i],rangeCds[i]))
```
