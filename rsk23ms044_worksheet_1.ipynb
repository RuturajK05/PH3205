```python
#Q1
import numpy as np
def is_prime(n):
    for i in range(2,int(np.sqrt(n))+1):
        if n%i==0: #checking remainder
            return False
            break
    return True
print(is_prime(16))
```

```python
#Q2
a = 5
b = 16
def gen():
    for i in range(a,b+1):
        if is_prime(i) == True:
            print(i)
gen()
#here, I have done it including both the given integers
```

```python
#Q3
def f(x,N):
    s=1
    b=1
    for i in range(1,N+1):
        b = b*(x/i)
        s +=b
    return s
f(1,50)
```

```python
#Q5
import matplotlib.pyplot as plt
import math

def Leibniz(n):
    s=1
    for i in range(1,n):
        i = ((-1)**i)*(1/((2*i)+1))
        s +=i
    print(4*s)
    return 4*s


def continuous(n):
    step = 1
    sum = 0
    while n > 0:
        if n == 1:
            step = 1/step
        else:
            step = 2 + (((2*n)-1)**2)/step
        n -= 1
    sum = 1 + step
    return 4/sum
    print(4/sum)

#creating empty lists
leibniz = []
cont_frac = []

#Printing outputs
x = [i for i in range(1,11)]
for i in x:
    print(f"{Leibniz(i):.8f}")
    leibniz.append(Leibniz(i))
    print(f"{continuous(i):.8f}")
    cont_frac.append(continuous(i))
    
#Plot
fig,ax = plt.subplots(figsize = (10,6))
ax.plot(x,leibniz, color = "green", ls = "-", label = "Leibniz")
ax.plot(x, cont_frac, color = "pink", ls = "-", label = "Continued Fraction")
plt.legend()
plt.show()
```
